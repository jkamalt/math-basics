# Предел функции <a class='anchor' id='func_limit'>

### Задание 1 <a class='anchor' id='func_limit_1'>

*Предложить пример функции, не имеющей предела в нуле и в бесконечностях.*

Функция 

$$f(x) = \frac{1}{\tan (x)}.$$

При $x \to \infty$ предел не существует.

В точке $x = 0$ есть только односторонние пределы и они не равны.

$$\lim_{x \to 0{-}} \frac{1}{\tan (x)} = -\infty,$$

$$\lim_{x \to 0{+}} \frac{1}{\tan (x)} = +\infty.$$

Пострим график, чтобы проиллюстрировать результат.

```python
x = np.linspace(-5, 5, 100)
y = 1 / np.tan(x)

plt.figure(figsize=(10, 10))
plt.axis([-5, 5, -10, 10])

plt.plot(x, y)
plt.grid()
plt.show()
```

Функция 

$$f(x) = \sin \frac{1}{x}.$$

При $x \to 0$ предел не существует.

Пострим график.

```python
x = np.linspace(-5, 5, 1000)
y = np.sin(1 / x)

plt.figure(figsize=(10, 10))

plt.plot(x, y)
plt.grid()
plt.show()
```

### Задание 3 <a class='anchor' id='func_limit_3'>

*Исследовать функцию*

$$f(x) = x^3 - x^2$$

*по плану*:
* *Область задания и область значений.*
* *Нули функции и их кратность.*
* *Отрезки знакопостоянства.*
* *Интервалы монотонности.*
* *Четность функции.*
* *Ограниченность.*
* *Периодичность.*

```python
x = np.linspace(-1.5, 1.5, 100)
y = np.power(x, 3) - np.power(x, 2)

plt.figure(figsize=(10, 10))

plt.plot(x, y)
plt.grid()
plt.show()
```

Очевидно, что область определения и область значений функции - это $\mathbb{R}$, т.е. вся числовая прямая.

Найдем **нули функции**. Для этого решим уравнение

$$x^3 - x^2 = 0,$$

$$x^2 (x - 1) = 0.$$

Корни этого уравнения:

$x_1 = 0$ - корень кратности 2,

$x_2 = 1$ - корень кратности 1.

Исследуем **знак функции** на интервалах $(-\infty, 0)$, $(0, 1)$, $(1, +\infty)$. Подставим в функцию значения из этих интервалов.

$$(-\infty, 0): f(-1) = -2,$$

$$(0, 1): f\left(\frac{1}{2}\right) = \frac{1}{8} - \frac{1}{4} < 0,$$

$$(1, +\infty): f(2) = 4,$$

<img src="func_sign.png">

Таким образом,

$(-\infty, 1)$ - функция отрицательная,

$(1, +\infty)$ - функция положительная.

Исследуем **монотонность** функции. Для этого вычислим производную функции.

$$f'(x) = 3 x^2 - 2x.$$

Найдем нули производной:

$$3 x^2 - 2x = 0,$$

$$x(3 x - 2) = 0.$$

Корни этого уравнения $x_1 = 0$, $x_2 = \frac{2}{3}$.

Исследуем знак производной на интервалах $(-\infty, 0)$, $\left(0, \frac{2}{3}\right)$, $\left(\frac{2}{3}, +\infty\right)$. Подставим в производную значения из этих интервалов.

$$(-\infty, 0): f'(-1) = 5,$$

$$\left(0, \frac{2}{3}\right): f'\left(\frac{1}{3}\right) = \frac{1}{3} - \frac{2}{3} = -\frac{1}{3},$$

$$\left(\frac{2}{3}, +\infty\right): f'(1) = 1.$$

<img src="der_sign.png">

Это также очевидно, т.к. это парабола, ветви которой направлены вверх и которая пересекает ось $y$ в двух точек. Таким образом, интервалы монотонности:

$(-\infty, 0)$ - функция возрастает,

$\left(0, \frac{2}{3}\right)$ - функция убывает,

$\left(\frac{2}{3}, +\infty\right)$ - функция возрастает.

Исследуем функцию на чётность.

$$f(x) = x^3 - x^2,$$

$$f(-x) = -x^3 - x^2 = - (x^3 + x^2) \neq - f(x) \neq f(x).$$

Т.е. это функция **общего вида**, что также видно из графика.

Очевидно, что функция **не ограничена** снизу и не ограничена сверху.

Исследуем функцию на периодичность.

$$f(x + T) = (x + T)^3 - (x + T)^2 = x^3 + 3 x^2 T + 3 x T^2 + T^3 -x^2 - 2 x T - T^2 = $$

$$x^3 - x^2(1 - 3T) + x(3 T^2 - 2T) + T^3 - T^2.$$

Очевидно, что это выражение равно $f(x) = x^3 - x^2$ только при $T = 0$, т.е. функция **не периодическая**. 

### Задание 4 <a class='anchor' id='func_limit_4'>

*Найти предел.*

#### Пример 1 <a class='anchor' id='func_limit_4_1'>

$$\lim_{x \to 0} \frac{3 x^3 - 2 x^2}{4 x^2}.$$

Вынесем $x^2$ за скобки:

$$\lim_{x \to 0} \frac{3 x^3 - 2 x^2}{4 x^2} = \lim_{x \to 0} \frac{x^2 (3 x - 2)}{4 x^2} = \lim_{x \to 0} \frac{3 x - 2}{4} = - \frac{1}{2}.$$

Получили

$$\lim_{x \to 0} \frac{3 x^3 - 2 x^2}{4 x^2} = - \frac{1}{2}.$$

#### Пример 2 <a class='anchor' id='func_limit_4_2'>

$$\lim_{x \to 0} \frac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1}.$$

Сделаем замену для удобства:

$$y = 1 + x.$$

Получим

$$\lim_{x \to 0} \frac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1} = \lim_{y \to 1} \frac{\sqrt{y} - 1}{\sqrt[3]{y} - 1}.$$

Умножим и разделим на сопряженные множители для числителя и знаменателя:

$$\lim_{y \to 1} \frac{\sqrt{y} - 1}{\sqrt[3]{y} - 1} = \lim_{y \to 1} \frac{\sqrt{y} - 1}{\sqrt[3]{y} - 1} \; \frac{\sqrt{y} + 1}{\sqrt{y} + 1} \; \frac{\sqrt[3]{y^2} + \sqrt[3]{y} + 1}{\sqrt[3]{y^2} + \sqrt[3]{y} + 1}.$$

Используя формулы сокращенного умножения получим в числителе и знаменателе:

$$\left(\sqrt{y} - 1\right) \left(\sqrt{y} + 1\right) = y - 1,$$

$$\left(\sqrt[3]{y} - 1\right) \left(\sqrt[3]{y^2} + \sqrt[3]{y} + 1\right) = y - 1.$$

Тогда

$$\lim_{y \to 1} \frac{\sqrt{y} - 1}{\sqrt[3]{y} - 1} = \lim_{y \to 1} \frac{y - 1}{y - 1} \; \frac{\sqrt[3]{y^2} + \sqrt[3]{y} + 1}{\sqrt{y} + 1} = \lim_{y \to 1} \frac{\sqrt[3]{y^2} + \sqrt[3]{y} + 1}{\sqrt{y} + 1} = \frac{3}{2}.$$

Получили

$$\lim_{x \to 0} \frac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1} = \frac{3}{2}.$$

# Теоремы о пределах <a class='anchor' id='limit_theorems'>

### Задание 1 <a class='anchor' id='limit_theorems_1'>

*Найти предел.*

#### Пример 1 <a class='anchor' id='limit_theorems_1_1'>

$$\lim_{x \to 0} \: \frac{\sin (2x)}{4x}.$$

В числителе представим $4x$ в виде $2 \cdot 2x$:

$$\lim_{x \to 0} \: \frac{\sin (2x)}{4x} = \lim_{x \to 0} \: \frac{1}{2} \frac{\sin (2x)}{2x} = \frac{1}{2} \; \lim_{x \to 0} \: \frac{\sin (2x)}{2x} = \frac{1}{2}.$$

Здесь использовали первый замечательный предел:

$$\lim_{x \to 0} \: \frac{\sin (2x)}{2x} = 1.$$

Получили

$$\lim_{x \to 0} \: \frac{\sin (2x)}{4x} = \frac{1}{2}.$$
