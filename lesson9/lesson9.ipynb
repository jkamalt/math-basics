### Задание 1 <a class='anchor' id='part_der_1'>

*Исследовать функцию на условный экстремум*

$$U = 3 - 8x + 6y,$$

*если*
    
$$x^2 + y^2 = 36.$$

Запишем функцию Лагранжа

$$L(\lambda, x, y) = U(x, y) + \lambda \varphi(x,y) = 3 - 8x + 6y + \lambda (x^2 + y^2 - 36).$$

Необходимые условия экстремума:

$$\begin{cases}
   L'_{x} = -8 + 2\lambda x = 0,\\
   L'_{y} = 6 + 2\lambda y = 0, \\
   L'_{\lambda} = x^2 + y^2 - 36 = 0.
\end{cases}$$

Выразим из первого уравнения $x$, а из второго $y$ и подставим в третье уравнение:

$$\begin{cases}
   x = 4\lambda,\\
   y = -3\lambda, \\
   16\lambda^2 + 9\lambda^2 - 36 = 0.
\end{cases}$$

Получим уравнение для $\lambda$

$$25\lambda^2 = 36.$$

Получили

$$\lambda = \pm \frac{6}{5}.$$

Подставим найденные значения $\lambda$ в выражения для $x$ и $y$ и получим две стационарные точки:

$$M_1\left(\frac{24}{5}; -\frac{18}{5} \right),$$

$$M_2\left(-\frac{24}{5}; \frac{18}{5} \right).$$

Исследуем эти точки на экстремум. Вычислим вторые производные функции Лагранжа

$$L''_{xx} = 2\lambda, \; L''_{yy} = 2\lambda, \; L''_{\lambda\lambda} = 0,$$

$$L''_{xy} = 0, \; L''_{x\lambda} = 2x, \; L''_{y\lambda} = 2y.$$

Вычислим гессиан:

$$\Delta = \begin{vmatrix}
    L''_{\lambda\lambda} & L''_{\lambda x} & L''_{\lambda y} \\
    L''_{x \lambda} & L''_{xx} & L''_{xy} \\
    L''_{y \lambda} & L''_{yx} & L''_{yy}
\end{vmatrix} = \begin{vmatrix}
    0 & 2x & 2y \\
    2x & 2\lambda & 0 \\
    2y & 0 & 2\lambda
\end{vmatrix} =  -2x 
\begin{vmatrix}
    2x & 0 \\
    2y & 2\lambda \\
\end{vmatrix} + 2y
\begin{vmatrix}
    2x & 2\lambda \\
    2y & 0 \\
\end{vmatrix} = -8\lambda x^2 - 8\lambda y^2 = -8\lambda (x^2 + y^2).$$

Вычислим гессиан для первой точки $M_1\left(\frac{24}{5}; -\frac{18}{5} \right)$ при $\lambda = \frac{6}{5}$:

$$\Delta_1 = -8 \frac{6}{5} (x^2 + y^2) < 0.$$

Т.е. точка $M_1\left(\frac{24}{5}; -\frac{18}{5} \right)$ - точка минимума.

Вычислим гессиан для второй точки $M_2\left(-\frac{24}{5}; \frac{18}{5} \right)$ при $\lambda = -\frac{6}{5}$:

$$\Delta_2 = 8 \frac{6}{5} (x^2 + y^2) > 0.$$

Т.е. точка $M_2\left(-\frac{24}{5}; \frac{18}{5} \right)$ - точка максимума.

Получили $M_1$, $M_2$ - точки экстремума функции $U = 3 - 8x + 6y$, а именно

$M_1\left(\frac{24}{5}; -\frac{18}{5} \right)$ - точка минимума,

$M_2\left(-\frac{24}{5}; \frac{18}{5} \right)$ - точка максимума.

### Задание 2 <a class='anchor' id='part_der_2'>

*Исследовать функцию на условный экстремум*

$$U = 2x^2 + 12xy + 32y^2 + 15,$$

*если*
    
$$x^2 + 16y^2 = 64.$$

Запишем функцию Лагранжа

$$L(\lambda, x, y) = U(x, y) + \lambda \varphi(x,y) = 2x^2 + 12xy + 32y^2 + 15 + \lambda (x^2 + 16y^2 - 64).$$

Необходимые условия экстремума:

$$\begin{cases}
   L'_{x} = 4x + 12y + 2\lambda x = 0,\\
   L'_{y} = 12x + 64y + 32\lambda y = 0, \\
   L'_{\lambda} = x^2 + 16y^2 - 64 = 0.
\end{cases}$$

Перепишем уравнения

$$\begin{cases}
   L'_{x} = x(2 + \lambda) + 6y = 0,\\
   L'_{y} = 3x + 8y(2 + \lambda) = 0, \\
   L'_{\lambda} = x^2 + 16y^2 - 64 = 0.
\end{cases}$$

Для простототы записи сделаем замену $a = 2 + \lambda$, получим

$$\begin{cases}
   ax + 6y = 0,\\
   3x + 8ay = 0, \\
   x^2 + 16y^2 - 64 = 0.
\end{cases}$$

Выразим из первого уравнения $x$, подставим в 2 и 3 уравнение и упростим:

$$\begin{cases}
   x = -\frac{6y}{a},\\
   -18y + 8a^2 y = 0, \\
   36y^2 + 16a^2 y^2 = 64a^2.
\end{cases}$$

Из 2 уравнения можно выразить $a$:

$$a^2 = \frac{18}{8} = \frac{9}{4},$$

$$a = \pm \frac{3}{2}.$$

Рассмотрим отдельно систему из 2 и 3 уравнения:

$$\begin{cases}
   -18y + 8a^2 y = 0, \\
   36y^2 + 16a^2 y^2 = 64a^2.
\end{cases}$$

Умножим 1 уравнение на $2y$:

$$\begin{cases}
   -36y^2 + 16a^2 y^2 = 0, \\
   36y^2 + 16a^2 y^2 = 64a^2.
\end{cases}$$

Сложим уравнение и получим

$$32a^2 y^2 = 64a^2,$$

$$y^2 = \frac{64}{32} = \frac{4}{2} = 2.$$

Получили

$$y = \pm \sqrt{2}.$$

Рассмотрим все варианты.

1) $a = \frac{3}{2}$

$$y = \sqrt{2} \; \Rightarrow \; x = -\frac{6y}{a} = - 6 \: \frac{2}{3} \: \sqrt{2} = - 4\sqrt{2} \; \Rightarrow M_1\left(- 4\sqrt{2}; \sqrt{2}\right).$$

$$y = -\sqrt{2} \; \Rightarrow \; x = -\frac{6y}{a} = 6 \: \frac{2}{3} \: \sqrt{2} = 4\sqrt{2} \; \Rightarrow M_2\left(4\sqrt{2}; -\sqrt{2}\right).$$

1) $a = -\frac{3}{2}$

$$y = \sqrt{2} \; \Rightarrow \; x = -\frac{6y}{a} = 6 \: \frac{2}{3} \: \sqrt{2} = 4\sqrt{2} \; \Rightarrow M_3\left(4\sqrt{2}; \sqrt{2}\right).$$

$$y = -\sqrt{2} \; \Rightarrow \; x = -\frac{6y}{a} = -6 \: \frac{2}{3} \: \sqrt{2} = -4\sqrt{2} \; \Rightarrow M_4\left(-4\sqrt{2}; -\sqrt{2}\right).$$

Получили 4 стационарные точки:

$$M_1\left(- 4\sqrt{2}; \sqrt{2}\right),$$

$$M_2\left(4\sqrt{2}; -\sqrt{2}\right),$$

$$M_3\left(4\sqrt{2}; \sqrt{2}\right),$$

$$M_4\left(-4\sqrt{2}; -\sqrt{2}\right).$$

Исследуем эти точки на экстремум. Вычислим вторые производные функции Лагранжа

$$L''_{xx} = 4 + 2\lambda, \; L''_{yy} = 64 + 32\lambda, \; L''_{\lambda\lambda} = 0,$$

$$L''_{xy} = 12, \; L''_{x\lambda} = 2x, \; L''_{y\lambda} = 32y.$$

Вычислим гессиан с учетом $a = 2 + \lambda$:

$$\Delta = \begin{vmatrix}
    L''_{\lambda\lambda} & L''_{\lambda x} & L''_{\lambda y} \\
    L''_{x \lambda} & L''_{xx} & L''_{xy} \\
    L''_{y \lambda} & L''_{yx} & L''_{yy}
\end{vmatrix} = \begin{vmatrix}
    0 & 2x & 32y \\
    2x & 4 + 2\lambda & 12 \\
    32y & 12 & 64 + 32\lambda
\end{vmatrix} =  8 \begin{vmatrix}
    0 & x & 16y \\
    x & a & 6 \\
    16y & 6 & 16a
\end{vmatrix} = -8x 
\begin{vmatrix}
    x & 6 \\
    16y & 16a \\
\end{vmatrix} + 128y
\begin{vmatrix}
    x & a \\
    16y & 6 \\
\end{vmatrix} = $$

$$ = -8x(16ax - 96y) + 128y(6x - 16ay) = -128ax^2 + 768xy + 768xy - 2048ay^2 = 1536xy - 128a(x^2 + 16y^2).$$

Получили

$$\Delta = 1536xy - 128a(x^2 + 16y^2).$$

1) Вычислим гессиан для первой точки $M_1\left(- 4\sqrt{2}; \sqrt{2}\right)$ при $a = \frac{3}{2}$:

$$\Delta_1 = -1536 \cdot 4\sqrt{2} \cdot \sqrt{2} - 128 \frac{3}{2}(x^2 + 16y^2) < 0.$$

Точка $M_1\left(- 4\sqrt{2}; \sqrt{2}\right)$ - точка минимума.

2) Вычислим гессиан для второй точки $M_2\left(4\sqrt{2}; -\sqrt{2}\right)$ при $a = \frac{3}{2}$:

$$\Delta_2 = -1536 \cdot 4\sqrt{2} \cdot \sqrt{2} - 128 \frac{3}{2}(x^2 + 16y^2) < 0.$$

Точка $M_2\left(4\sqrt{2}; -\sqrt{2}\right)$ - точка минимума.

3) Вычислим гессиан для третьей точки $M_3\left(4\sqrt{2}; \sqrt{2}\right)$ при $a = -\frac{3}{2}$:

$$\Delta_3 = 1536 \cdot 4\sqrt{2} \cdot \sqrt{2} + 128 \frac{3}{2}(x^2 + 16y^2) > 0.$$

Точка $M_3\left(4\sqrt{2}; \sqrt{2}\right)$ - точка максимума.

4) Вычислим гессиан для четвертой точки $M_4\left(-4\sqrt{2}; -\sqrt{2}\right)$ при $a = -\frac{3}{2}$:

$$\Delta_4 = 1536 \cdot 4\sqrt{2} \cdot \sqrt{2} + 128 \frac{3}{2}(x^2 + 16y^2) > 0.$$

Точка $M_4\left(-4\sqrt{2}; -\sqrt{2}\right)$ - точка максимума.

Получили $M_1$, $M_2$, $M_3$, $M_4$ - точки экстремума функции $U = 2x^2 + 12xy + 32y^2 + 15$, а именно

$M_1\left(- 4\sqrt{2}; \sqrt{2}\right)$ - точка минимума,

$M_2\left(4\sqrt{2}; -\sqrt{2}\right)$ - точка минимума,

$M_3\left(4\sqrt{2}; \sqrt{2}\right)$ - точка максимума,

$M_4\left(-4\sqrt{2}; -\sqrt{2}\right)$ - точка максимума.

### Задание 3 <a class='anchor' id='part_der_3'>

*Найти производную функции*

$$U = x^2 + y^2 + z^2$$

*по направлению вектора*

$$\vec{c}(-9, 8, -12)$$
    
*в точке* $M(8; -12; 9)$. 

Производная функции $U$ по направлению вектора $\vec{c}$ в точке $M$ вычисляется по формуле:

$$\left. \frac{\partial U}{\partial c} \right|_{M} = \left. \frac{\partial U}{\partial x} \right|_{M} \cos \alpha + \left. \frac{\partial U}{\partial y} \right|_{M} \cos \beta + \left. \frac{\partial U}{\partial z} \right|_{M} \cos \gamma,$$

где $\cos \alpha$, $\cos \beta$, $\cos \gamma$ - координаты векторы единичной длины, определяющего направление вектора $\vec{c}$.

Запишем частные производные

$$\frac{\partial U}{\partial x} = 2x,$$

$$\frac{\partial U}{\partial y} = 2y,$$

$$\frac{\partial U}{\partial z} = 2z.$$

Вычислим частные производные в точке $M(8; -12; 9)$:

$$\left. \frac{\partial U}{\partial x} \right|_{M} = 16,$$

$$\left. \frac{\partial U}{\partial y} \right|_{M} = -24,$$

$$\left. \frac{\partial U}{\partial z} \right|_{M} = 18.$$

Найдем координаты направляющего вектора единичной длины. Вычислим длину вектора $\vec{c}$:

$$|\vec{c}| = \sqrt{81 + 64 + 144} = 17.$$

Тогда координаты направляющего вектора единичной длины можно записать как

$$\vec{c}_0 = \left(-\frac{9}{17}, \frac{8}{17}, -\frac{12}{17} \right).$$

Тогда производная функции $U$ по направлению вектора $\vec{c}$ в точке $M$ равна

$$\left. \frac{\partial U}{\partial c} \right|_{M} = -16\frac{9}{17} - 24\frac{8}{17} - 18\frac{12}{17} = -\frac{552}{17}.$$

Получили

$$\left. \frac{\partial U}{\partial c} \right|_{M} = -\frac{552}{17}.$$

### Задание 4 <a class='anchor' id='part_der_4'>

*Найти производную функции*

$$U = e^{x^2 + y^2 + z^2}$$

*по направлению вектора*

$$\vec{d}(4, -13, -16)$$
    
*в точке* $L(-16; 4; -13)$. 

Производная функции $U$ по направлению вектора $\vec{d}$ в точке $L$ вычисляется по формуле:

$$\left. \frac{\partial U}{\partial d} \right|_{L} = \left. \frac{\partial U}{\partial x} \right|_{L} \cos \alpha + \left. \frac{\partial U}{\partial y} \right|_{L} \cos \beta + \left. \frac{\partial U}{\partial z} \right|_{L} \cos \gamma,$$

где $\cos \alpha$, $\cos \beta$, $\cos \gamma$ - координаты векторы единичной длины, определяющего направление вектора $\vec{d}$.

Запишем частные производные

$$\frac{\partial U}{\partial x} = 2x e^{x^2 + y^2 + z^2},$$

$$\frac{\partial U}{\partial y} = 2y e^{x^2 + y^2 + z^2},$$

$$\frac{\partial U}{\partial z} = 2z e^{x^2 + y^2 + z^2}.$$

Вычислим частные производные в точке $L(-16; 4; -13)$:

$$\left. \frac{\partial U}{\partial x} \right|_{L} = -32e^{441},$$

$$\left. \frac{\partial U}{\partial y} \right|_{L} = 8e^{441},$$

$$\left. \frac{\partial U}{\partial z} \right|_{L} = -26e^{441}.$$

Найдем координаты направляющего вектора единичной длины. Вычислим длину вектора $\vec{d}$:

$$|\vec{d}| = \sqrt{16 + 169 + 256} = 21.$$

Тогда координаты направляющего вектора единичной длины можно записать как

$$\vec{d}_0 = \left(\frac{4}{21}, -\frac{13}{21}, -\frac{16}{21} \right).$$

Тогда производная функции $U$ по направлению вектора $\vec{d}$ в точке $M$ равна

$$\left. \frac{\partial U}{\partial d} \right|_{L} = -32e^{441}\frac{4}{21} - 8e^{441}\frac{13}{21} + 26e^{441}\frac{16}{21} = e^{441}\frac{184}{21}.$$

Получили

$$\left. \frac{\partial U}{\partial d} \right|_{L} = e^{441}\frac{184}{21}.$$
