# Понятие о производной <a class='anchor' id='der'>

### Задание 1 <a class='anchor' id='der_1'>

*Найти длину $x$ и ширину $y$ прямоугольника при заданном периметре $P$=144 cм, при которых данный прямоугольник имеет наибольшую площадь $S$.*

Функция, для которой нужно найти максимальное значение - площадь прямоугольника

$$S = xy.$$

Выразим ширину прямоугольника $y$ через длину $x$, используя равенство для периметра:

$$x + y = P,$$

$$y = P - x.$$

Подставим в выражение для площади:

$$S = xy = x(P - x) = -x^2 + Px.$$

Найдем точки экстремума. Вычислим производную:

$$S' = -2x + P.$$

Решим уравнение:

$$S' = -2x + P = 0,$$

$$x = \frac{P}{2}.$$

Проверим, что найденная точка - точка максимума:

$$S'(0) = P > 0,$$

$$S'(P) = -2P + P = -P < 0.$$

Т.е. при переходе через точку $x = \frac{P}{2}$ производная $S'$ меняет свой знак с $+$ на $-$.

Полученные значение длины и ширины, при которых прямоугольник с заданным периметром $P$ имеет наибольшую площадь:

$$x = \frac{P}{2},$$

$$y = P - \frac{P}{2} = \frac{P}{2}.$$

Тогда для значения $P = 144$ получим:

$$x = y = 72.$$

### Задание 2 <a class='anchor' id='der_2'>

*Найти экстремумы функций, если они есть.*

#### Пример 1 <a class='anchor' id='der_2_1'>

$$y = |2x|.$$

Вычислим производную:

$$y' = 2 \frac{x}{|x|}.$$

Найдем критические точки. В точке $x = 0$ производная не существует. Также нет таких точек, где производная равна 0.

Очевидно, что при переходе через точку $x = 0$ производная меняет знак с $-$ на $+$, т.е. эта точка - точка экстремума, точнее точка минимума.

Получили $x = 0$ - **точка экстремума** функции $y = |2x|$, а именно точка минимума.

```python
x = np.linspace(-5, 5, 1000)
y = np.abs(2 * x)
y_der = 2 * x / np.abs(x)

plt.figure(figsize=(10, 10))

plt.plot(x, y, label='y')
plt.plot(x, y_der, label="y'")
plt.grid()
plt.legend()
plt.show()
```

#### Пример 2 <a class='anchor' id='der_2_2'>

$$y = x^3.$$

Вычислим производную:

$$y' = 3x^2.$$

Найдем критические точки. Решим уравнение:

$$y' = 3x^2 = 0,$$

$$x = 0.$$

Очевидно, что при переходе черех точку $x=0$ производная не меняет знак, т.е. это не точка экстремума. Это точка перегиба.

Таким образом, функция $y = x^3$ **не имеет точек экстремума**, но имеет одну критическую точку $x = 0$.

```python
x = np.linspace(-5, 5, 100)
y = np.power(x, 3)
y_der = 3 * np.power(x, 2)

plt.figure(figsize=(10, 10))

plt.plot(x, y, label='y')
plt.plot(x, y_der, label="y'")
plt.grid()
plt.legend()
plt.show()
```

#### Пример 3 <a class='anchor' id='der_2_3'>

$$y = e^{3x}.$$

Выислим производную

$$y' = 3 e^{3x}.$$

Очевидно, что для такой производной нет точек, где она не существует или равна 0.

Таким образом, функция $y = e^{3x}$ **не имеет критических точек**.

```python
x = np.linspace(-1, 1, 100)
y = np.exp(3 * x)
y_der = 3 * np.exp(3 * x)

plt.figure(figsize=(10, 10))

plt.plot(x, y, label='y')
plt.plot(x, y_der, label="y'")
plt.grid()
plt.legend()
plt.show()
```

#### Пример 4 <a class='anchor' id='der_2_4'>

$$y = x^3 - 5x.$$

Вычислим производную:

$$y' = 3x^2 - 5.$$

Найдем критические точки. Решим уравнение:

$$y' = 3x^2 - 5 = 0,$$

$$x_1 = -\sqrt{\frac{5}{3}}, \; x_2 = \sqrt{\frac{5}{3}}.$$

Исследуем знак производной на интервалах $\left(-\infty, -\sqrt{\frac{5}{3}}\right)$, $\left(-\sqrt{\frac{5}{3}}, \sqrt{\frac{5}{3}}\right)$, $\left(\sqrt{\frac{5}{3}}, +\infty\right)$. Подставляя значения из интервалов получим такие знаки производной:

$$\left(-\infty, -\sqrt{\frac{5}{3}}\right): y'\left(-\sqrt{5}\right) = 10,$$

$$\left(-\sqrt{\frac{5}{3}}, \sqrt{\frac{5}{3}}\right): y'(0) = -5,$$

$$\left(\sqrt{\frac{5}{3}}, +\infty\right): y'\left(\sqrt{5}\right) = 10.$$

<img src="der_sign.png">

Получили $x_1$, $x_2$ - **точки экстремума** функции $y = x^3 - 5x$, а именно

$x_1 = -\sqrt{\frac{5}{3}}$ - точка максимума,

$x_2 = \sqrt{\frac{5}{3}}$ - точка минимума.

```python
x = np.linspace(-5, 5, 100)
y = np.power(x, 3) - 5 * x
y_der = 3 * np.power(x, 2) - 5

plt.figure(figsize=(10, 10))

plt.plot(x, y, label='y')
plt.plot(x, y_der, label="y'")
plt.grid()
plt.legend()
plt.show()
```
